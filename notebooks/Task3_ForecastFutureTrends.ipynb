```python
import os
import sys


sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

# notebooks/Task3_ForecastFutureTrends.ipynb

from scripts.forecasting_utils import (
    load_tesla_data,
    split_train_test,
    train_arima_model,
    forecast_future_arima,
    plot_future_forecast
)

# Step 1: Load Tesla's closing prices
tesla_close = load_tesla_data()

# Step 2: Split into training and testing sets
train, test = split_train_test(tesla_close)

# Step 3: Train the ARIMA model
arima_model = train_arima_model(train)

# Step 4: Forecast future values
forecast, conf_int = forecast_future_arima(arima_model, steps=180)

# Step 5: Visualize the results
plot_future_forecast(train, test, forecast, conf_int)
```
