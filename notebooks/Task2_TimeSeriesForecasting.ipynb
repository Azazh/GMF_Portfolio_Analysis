```python
import sys
import os
import pandas as pd

# Add the project root directory to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))

# notebooks/Task2_TimeSeriesForecasting.ipynb

from scripts.forecasting_utils import (
    load_tesla_data,
    split_train_test,
    train_arima_model,
    forecast_arima,
    evaluate_model,
    plot_forecast
)
from scripts.lstm_utils import prepare_lstm_data, build_lstm_model

# Step 1: Load Tesla's closing prices
tesla_close = load_tesla_data()

# Step 2: Split into training and testing sets
train, test = split_train_test(tesla_close)

# Step 3: Train the ARIMA model
arima_model = train_arima_model(train)

# Step 4: Forecast future values
forecast = forecast_arima(arima_model, test)

# Step 5: Evaluate the model
evaluate_model(test, forecast)

# Step 6: Visualize the results
plot_forecast(test, forecast)
```

```python
# notebooks/Task2_TimeSeriesForecasting.ipynb (continued)
# Step 1: Prepare data for LSTM
X, y, scaler = prepare_lstm_data(tesla_close)

# Step 2: Split into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Step 3: Build and train the LSTM model
lstm_model = build_lstm_model(input_shape=(X_train.shape[1], 1))
lstm_model.fit(X_train, y_train, batch_size=32, epochs=10)

# Step 4: Forecast future values
predictions = lstm_model.predict(X_test)
predictions = scaler.inverse_transform(predictions.reshape(-1, 1))

# Step 5: Evaluate the model
evaluate_model(y_test, predictions)

# Step 6: Visualize the results
plot_forecast(pd.Series(y_test.flatten()), pd.Series(predictions.flatten()))
```
